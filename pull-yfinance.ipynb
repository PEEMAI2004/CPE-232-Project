```python
import pandas as pd
import os

output_folder = "/home/kamin/Documents/CPE 241 Project/sp500_data"

# Assuming output_folder is already defined
dataframes = {}
for file_name in os.listdir(output_folder):
    if file_name.endswith('.csv'):
        ticker = file_name.split('.')[0]
        file_path = os.path.join(output_folder, file_name)
        dataframes[ticker] = pd.read_csv(file_path)

# Combine all DataFrames into a single 3D DataFrame
combined_df = pd.concat(dataframes, axis=0, keys=dataframes.keys())

# Print the first few rows of the combined DataFrame
combined_df
```
